HTTP response 처리하기
- 1. response 객체의 이해
- 2. status_code 확인하기
- 3. text 속성 확인하기

```python
url = 'https://news.v.daum.net/v/20190728165812603'
resp = requests.get(url)
if resp.status_code  == 200:
    print(resp.headers)
else :
    print('error')
```
